```python
import numpy as np
import networkx as nx
import matplotlib.pyplot as plt
```

```python
links = 200
nodes = 100
```

```python
def creating_adjacency_matrix(nodes, links):
    num_link = 0
    L = np.zeros((nodes, nodes))
    while num_link < links:
        i = np.random.randint(nodes)
        j = np.random.randint(nodes)
        if (i != j) and L[i, j] == 0:
            L[i, j] = 1
            L[j, i] = 1
            num_link += 1
    return L
```

```python
L = creating_adjacency_matrix(nodes, links)
```

```python
np.sum(L == 1) / 2
```

```python
G = nx.from_numpy_matrix(L)
plt.figure(figsize=(12, 10))
plt.title('Graph')
nx.draw_networkx(G)
plt.savefig('../Pictures/task5_graph', bbox_inches='tight')
```

```python
def print_eges(adjlist):
    for line in adjlist:
        print(line)
print_eges(nx.generate_adjlist(G))
```
