```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
fig = plt.figure(figsize=(15, 10))
plt.plot(np.arange(0, 10, 0.1), np.exp(-np.arange(0, 10, 0.1)),label="Exponenta")
plt.legend()
# plt.grid()
plt.xlabel("t")
plt.ylabel("S1")
plt.savefig('graphs.jpg', bbox_inches='tight')
plt.show()
```
